#### 15. Create a 2d array with 1 on the border and 0 inside (★☆☆)

```python
def create_bordered_array_numpy(rows, cols):
    """
    Creates a 2D NumPy array of a given size with 1s on the border and 0s inside.
    """
    # Create an array of all ones with the specified shape (rows, cols)
    array = np.ones((rows, cols), dtype=int)
    
    # Use slicing to set the inner part of the array to 0
    # [1:-1, 1:-1] selects from the second row to the second-to-last row,
    # and from the second column to the second-to-last column.
    if rows > 2 and cols > 2:
        array[1:-1, 1:-1] = 0
        
    return array

# Example for a 5x5 array
bordered_array_numpy = create_bordered_array_numpy(5, 5)
print(bordered_array_numpy)
```
