Easy Referencing of Numpy Library

```python
import numpy as np
```

Sorting Arrays

```python
a = np.array ([1,2,4,3,5])
np.sort(a)
```

```python
# Default - Sort along rows.(axis=-1)
a = np.array ([[23,15,7],[11,19,78],[3,89,73]])
np.sort(a)
```

```python
# Sort everything
np.sort(a, axis=None)
```

```python
# Sort along column
np.sort(a, axis=1)
```

```python
# Sort
np.sort(a, axis=0)
```

Indices of the elements

```python
b = np.array ([23,15,20])
# np.sort(b)
np.argsort(b)
```

Partitioning

```python
# Partial Sort or Creating a wall
b
```

```python
# Elements to the right of k are less and to the right are greater.
np.partition(b,2)   
```
